```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize = (7,5))
```

```python
# 正选曲线   
import numpy as np
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)
plt.plot(x,y,'bp--')
plt.show()
```

```python
# 饼图    plt.pie(size)
import matplotlib.pyplot as plt
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
size = [15,30,45,10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)   # 突出显示第二块
plt.pie(size, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True, startangle=90)
plt.axis('equal')     # 显示为圆，避免压缩为椭圆
plt.show()
```

```python
# 条形图    plt.hist()
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randn(1000)   # 生成1000个符合正太分布的数
plt.hist(x,10)    # 分成十组进行绘制条形图
plt.show()
```

```python
# 绘制箱型图    D.plot(kind = 'box')     D.boxplot()   
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.random.randn(1000)
D = pd.DataFrame([x, x+1]).T    # 构造两列的DataFrame
D.plot(kind = 'box')
plt.show()
```

```python
# 绘制对数函数   plot(logy = True)      plot(logx = True)
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize = (7,5))
import numpy as np
import pandas as pd
```

```python
x = pd.Series(np.exp(np.arange(20)))
x.plot(label = u'原始数据', legend = True)    
plt.show()
```

```python
x.plot(logy = True, label = u'对数数据图', legend = True )
plt.show()
```

```python
# 绘制误差条形图   y.plot(yerr = error)
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize = (7,5))
import numpy as np
import pandas as pd

error = np.random.randn(10)
y = pd.Series(np.sin(np.arange(10)))     # 均值数据列
y.plot(yerr = error)
plt.show()
```
