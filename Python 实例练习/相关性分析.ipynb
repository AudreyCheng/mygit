```python
#餐饮销量数据相关性分析
from __future__ import print_function
import pandas as pd
```

```python
catering_sale = '/Users/admin/PycharmProjects/pythonProject1/python_da/chapter3/demo/data/catering_sale_all.xls' #餐饮数据，含有其他属性
data = pd.read_excel(catering_sale, index_col = u'日期') #读取数据，指定“日期”列为索引列
data.head()    # 显示表的前5行
```

```python
data.corr()      # 方法一，相关系数矩阵，即给出了任意两款菜式之间的相关系数
```

```python
data.corr()[u'百合酱蒸凤爪']  #只显示“百合酱蒸凤爪”与其他菜式的相关系数
```

```python
data[u'百合酱蒸凤爪'].corr(data[u'翡翠蒸香茜饺']) #计算“百合酱蒸凤爪”与“翡翠蒸香茜饺”的相关系数
```

```python
data.sum()
```

```python
data.sum()[u'香煎韭菜饺']     # 计算想剪韭菜饺的所有列 
```

```python
data.mean()[u'香煎韭菜饺']    # 计算香煎韭菜饺的所有列的平均值
```

```python
data.var()    # 计算样本的方差
```

```python
data.std()   # 计算样本的标准差
```

```python
data.cov()    # 计算样本的协方差矩阵
```

```python
data.skew()   # 计算样本的偏度 （三阶矩）
```

```python
data.kurt()   # 计算样本的峰度 （四阶矩）
```
