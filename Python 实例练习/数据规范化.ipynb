```python
#-*- coding: utf-8 -*-
#数据规范化
import pandas as pd
import numpy as np
```

```python
datafile = '/Users/admin/PycharmProjects/pythonProject1/python_da/chapter4/demo/data/normalization_data.xls'
data = pd.read_excel(datafile, header = None)    # 读取数据
```

```python
a = ((data - data.min())/(data.max()-data.min()))       # 最小 —— 最大规范化
b = ((data - data.mean()))/ data.std()                  # 零 —— 均值规范化
c = data/10**np.log10(data.abs().max())                 # 小数定标规范化
```
