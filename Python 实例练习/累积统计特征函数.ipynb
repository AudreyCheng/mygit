```python
#餐饮销量数据相关性分析
from __future__ import print_function
import pandas as pd
```

```python
catering_sale = '/Users/admin/PycharmProjects/pythonProject1/python_da/chapter3/demo/data/catering_sale_all.xls' 
data = pd.read_excel(catering_sale, index_col=u'日期')
data.head()
```

```python
cumdata = data.cumsum()     # 依次给出前1、2、3。。。行的和
cumdata.head()
```

```python
cumdata1 = data.cumprod()     # 依次给出前1、2、3。。。行的积
cumdata1.head()
```

```python
maxdata = data.cummax()
maxdata.head()
```

```python
maxdata1 = data.cummin()
maxdata1.head()
```
